```python
# Powerplay (overs 1 to 6 for both innings)
final_df['Powerplay_Scores'] = final_df[[f'innings_1_over_{i}' for i in range(1, 7)] +
                                        [f'innings_2_over_{i}' for i in range(1, 7)]].sum(axis=1)

# Middle Overs (overs 7 to 15 for both innings)
final_df['Middle_Overs_Scores'] = final_df[[f'innings_1_over_{i}' for i in range(7, 16)] +
                                            [f'innings_2_over_{i}' for i in range(7, 16)]].sum(axis=1)

# Death Overs (overs 16 to 20 for both innings)
final_df['Death_Overs_Scores'] = final_df[[f'innings_1_over_{i}' for i in range(16, 21)] +
                                           [f'innings_2_over_{i}' for i in range(16, 21)]].sum(axis=1)
```

```python
import numpy as np

# Conditions
condition_wickets = (final_df['second_ings_score'] > final_df['first_ings_score'])
condition_runs = (final_df['first_ings_score'] > final_df['second_ings_score'])
condition_draw = final_df['first_ings_score'].isna() | final_df['second_ings_score'].isna()

# Apply conditions
final_df['Win_Type'] = np.select(
    [condition_wickets, condition_runs, condition_draw],
    ['wickets', 'runs', 'draw'],
    default='draw'  # fallback if none of the above match
)
```

```python
import numpy as np

# Create Win_Margin column
final_df['Win_Margin'] = np.where(
    final_df['Win_Type'] == 'runs',
    final_df['first_ings_score'] - final_df['second_ings_score'],
    np.where(
        final_df['Win_Type'] == 'wickets',
        final_df['second_ings_score'] - final_df['first_ings_score'],
        0
    )
)

# Optional: Convert to integer
final_df['Win_Margin'] = final_df['Win_Margin'].astype(int)
```
