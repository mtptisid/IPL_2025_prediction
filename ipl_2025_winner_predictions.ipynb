```python
# Function to compute over number
def compute_over_number(ball_no):
    base = int(ball_no)
    if abs(ball_no - base) < 1e-6:
        return base
    return base + 1
```

```python
df['Over_Number'] = df['Ball No'].apply(compute_over_number)

# Preview
print(df[['Ball No', 'Over_Number']].head(20))
```

```python
df.head(15)
```

```python
grouped = df.groupby([
    'Match id', 'Date', 'Season', 'Batting team', 'Bowling team', 'Over_Number', 'Innings No'
], as_index=False).agg({
    'runs_scored': 'sum',
    'extras': 'sum',
    'score': 'max'
})
```

```python
grouped['Innings No'].unique()
```

```python
grouped
```

```python
# Apply transformation
grouped['Innings No'] = np.where(
    grouped['Innings No'] > 2,
    np.where(grouped['Innings No'] % 2 == 1, 1, 2),
    grouped['Innings No']
)
```

```python
import pandas as pd

# Step 1: Get consistent team order
team_order = grouped.groupby("Match id").first().reset_index()
team_order["Team1"] = team_order["Batting team"]
team_order["Team2"] = team_order["Bowling team"]
team_order = team_order[["Match id", "Team1", "Team2"]]

# Step 2: Create over-wise column names
grouped["over_column"] = grouped.apply(
    lambda row: f"innings_{int(row['Innings No'])}_over_{int(row['Over_Number'])}", axis=1
)

# Step 3: Pivot the table
pivot_df = grouped.pivot_table(
    index=["Match id", "Date", "Season"],
    columns="over_column",
    values="runs_scored",
    aggfunc="sum"
).reset_index()

# Step 4: Flatten multi-level column names if needed
pivot_df.columns.name = None
pivot_df.columns = [str(col) for col in pivot_df.columns]

# Step 5: Merge back with team info
final_df = pd.merge(pivot_df, team_order, on="Match id")

# Step 6: Sort the over columns numerically
def sort_key(col):
    if col.startswith("innings_"):
        parts = col.split("_")
        return (int(parts[1]), int(parts[3]))
    return (-1, -1)  # ensures 'Match id', 'Date' etc. come first

sorted_over_columns = sorted([col for col in final_df.columns if col.startswith("innings_")], key=sort_key)

# Step 7: Reorder all columns
cols = ['Match id', 'Date', 'Season', 'Team1', 'Team2'] + sorted_over_columns
final_df = final_df[cols]
```

```python
# Convert Match id columns to sets
final_ids = set(final_df['Match id'])
df2_ids = set(df2['Match_ID'])

# Find missing Match ids
missing_ids = final_ids - df2_ids  # present in final_df but not in df2

print("Missing Match ids:", missing_ids)
print("Number of missing Match ids:", len(missing_ids))
```

```python
missing_rows = final_df[final_df['Match id'].isin(missing_ids)]
missing_rows.head()
```

```python
# Fill all NaN values with 0
final_df = final_df.fillna(0)

# Drop innings_1_over_0 and innings_2_over_0 if they exist
columns_to_drop = ['Team1', 'Team2','innings_1_over_0', 'innings_2_over_0',	'Date',	'Season']
final_df = final_df.drop(columns=[col for col in columns_to_drop if col in final_df.columns])
```
